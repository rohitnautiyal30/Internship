```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.flipkart.com/fire-boltt-cyclone-pro-1-43-amoled-display-466-466-px-resolution-123-sports-smartwatch/p/itm564c9582ba328?pid=SMWGZRY27GD3QF8V&lid=LSTSMWGZRY27GD3QF8VXAHGJW"

page = requests.get(url)

page
```

```python
print(page.content)
```

```python
soup = BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

### getting user names from the reviews

```python
names = soup.find_all('p',class_="_2sc7ZR _2V5EHH")
names
```

```python
len(names)
```

```python
cust_name = []

for i in names:
    cust_name.append(i.get_text())
cust_name
```

### getting review titles

```python
titles = soup.find_all('p',class_="_2-N8zT")

titles
```

```python
len(titles)
```

```python
review_titles = []

for i in titles :
    review_titles.append(i.text)
review_titles
```

```python
len(review_titles)
```

```python
ratings = soup.find_all('div', class_="_3LWZlK _1BLPMq")

ratings
```

### getting review ratings

```python
len(ratings)
```

```python
review_ratings = []

for i in ratings :
    review_ratings.append(i.get_text())
review_ratings
```

### getting review body

```python
review = soup.find_all('div', class_="t-ZTKy")

review
```

```python
len(review)
```

```python
review_body =[]

for i in review :
    review_body.append(i.get_text())
    
review_body
```

### saving data into a dataframe

```python
import pandas as pd
```

```python
User_reviews = pd.DataFrame({'Name': cust_name, 
                             'Review Titles': review_titles, 
                        
                            'Reviews':review_body})

User_reviews
```
